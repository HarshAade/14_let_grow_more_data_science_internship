        01.4) Image to Pencil Sketch with Python


```python
# Importing Required Libraries
from google.colab.patches import cv2_imshow
import cv2
```

```python
# Loading image
img = cv2.imread('ganpati.jpeg')
```

```python
cv2_imshow(img)
```

```python
# Converting an image into Gray Sacle Image
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_grey)
```

```python
# Inverting the image
img_ivert = cv2.bitwise_not(img_grey)
```

```python
cv2_imshow(img_ivert)
```

```python
# Smoothing the image
img_smoothig = cv2.GaussianBlur(img_ivert,(21,21),sigmaX=0,sigmaY=0)
```

```python
cv2_imshow(img_smoothig)
```

```python
# Obtaining the final Sketch
def dodgeV2(x,y):
  return cv2.divide(x,255-y,scale=256)
```

```python
# Final outpu as penical
final_img = dodgeV2(img_grey,img_smoothig)
```

```python
cv2_imshow(final_img)
```
